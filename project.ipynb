```python
import numpy as np
import pandas as pd
```

```python
datas = pd.read_csv("global_disaster_response_2018_2024 (1).csv")
df = pd.DataFrame(datas)
df
```

```python
df.isna().sum()
```

```python
df.drop(columns="date",inplace=True)
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df["country"] = le.fit_transform(df["country"])
df["disaster_type"] = le.fit_transform(df["disaster_type"])
```

```python
df
```

```python
X = df.iloc[:,df.columns != "recovery_days"]
X
```

```python
y = df.iloc[:,df.columns == "recovery_days"]
y
```

```python
from matplotlib import pyplot 

# Load your dataset # Create a 2×2 subplot grid

pyplot.figure(figsize=(12, 10)) 

# 1. Severity vs Recovery Days 
pyplot.subplot(2, 2, 1) 
pyplot.scatter(df['severity_index'], df['recovery_days']) 
pyplot.xlabel("Severity Index") 
pyplot.ylabel("Recovery Days") 
pyplot.title("Severity vs Recovery Days") 

# 2. Economic Loss vs Recovery Days
pyplot.subplot(2, 2, 2)
pyplot.scatter(df['economic_loss_usd'], df['recovery_days']) 
pyplot.xlabel("Economic Loss (USD)") 
pyplot.ylabel("Recovery Days")
pyplot.title("Economic Loss vs Recovery Days") 

# 3. Response Time vs Recovery Days 
pyplot.subplot(2, 2, 3) 
pyplot.scatter(df['response_time_hours'], df['recovery_days']) 
pyplot.xlabel("Response Time (Hours)") 
pyplot.ylabel("Recovery Days") 
pyplot.title("Response Time vs Recovery Days") 

# 4. Aid Amount vs Recovery Days 
pyplot.subplot(2, 2, 4)
pyplot.scatter(df['aid_amount_usd'], df['recovery_days']) 
pyplot.xlabel("Aid Amount (USD)")
pyplot.ylabel("Recovery Days") 
pyplot.title("Aid Amount vs Recovery Days") 
pyplot.tight_layout() 
pyplot.show()

```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
y_test
from sklearn.metrics import mean_absolute_percentage_error
mape = mean_absolute_percentage_error(y_pred,y_test)
mape
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_pred,y_test)
r2
```
